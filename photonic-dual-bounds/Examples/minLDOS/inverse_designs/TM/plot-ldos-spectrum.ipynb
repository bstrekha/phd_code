```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = 'TMLDOSmin4+0j_dwQabs1e-1to1e7_periodicfullspace10by10_period1_maxeval500_Wfunc_normalized_vacstart_Qabs5.0e+06_L10.0_optdof_spectrum.txt'
ldos_list = np.loadtxt(filename)
omegas = np.linspace(-1e6/2/10**(6),1e6/2/10**(6), len(ldos_list)) + 1
plt.plot(omegas, ldos_list, color='#7f7f7f',linewidth=2.5)
plt.xlabel('$\omega/\omega_{0}$')
plt.ylabel('$\\rho(\omega)/\\rho_{vac}(\omega_{0})$')
plt.yscale('log')
# plt.savefig('spectrum.png', bbox_inches='tight')
# plt.savefig('spectrum.pdf', bbox_inches='tight')
```
