```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
design_file = 'results/cloak_1by1_chif4+1e-3j_chid2+1e-3j_Qabs5.0e+05_L1.0_optdof.npy'
design = np.load(design_file)

background = np.zeros_like(design)
background_file = 'results/cloak_1by1_chif4+1e-3j_chid2+1e-3j_L1.0_backgrounddof.npy'
background = np.load(background_file)
plt.figure(figsize=((7,7)))
plt.imshow(design + background/2,cmap='Greys')
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.savefig('cloak_design.png', bbox_inches='tight')
```
